```python
## Shared Utility Functions
#  Schema validator 
def validate_schema(df, required_columns):
    missing = [c for c in required_columns if c not in df.columns]
    if missing:
        raise ValueError(f"required columns {missing} missing")
    print(f"schema OK, {required_columns} present")
    return True

#  Row counter
def log_counter(df, name):
    counter = df.count()
    print(f"INFO : {counter} records in {name}")

#  Null checker 
from pyspark.sql.functions import col, sum as sum_check
def null_checker(df):
   null_counts = df.select([sum_check(col(c).isNull().cast("int")).alias(c) for c in df.columns])
   null_counts.show()

print("INFO: Utility functions Loaded")

```
