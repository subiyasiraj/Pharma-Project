```python
#  Cell 2: Read CSV from DBFS 

df_raw = spark.read.option("header","true") \
    .option("inferSchema", "true") \
    .csv(CSV_PATH)

log_counter(df_raw, "df_raw")
```

```python
#  Cell 3: Validate schema
required = ['sale_id','sale_date','drug_name','category',
'region','units_sold','unit_price','total_amount']
validate_schema(df_raw, required)
null_checker(df_raw)
```

```python
#  Cell 5: Display + schema
df_raw.display()
df_raw.printSchema()
```

```python
# Cell 6: Register temp view 

df_raw.createOrReplaceTempView("sales_raw")
print("Temp view raw sales created")
```

```python
%sql
SELECT region,
COUNT(*) AS total_records,
SUM(units_sold) AS total_units,
ROUND(SUM(total_amount),2) AS revenue
FROM sales_raw
GROUP BY region
ORDER BY revenue DESC;
```
